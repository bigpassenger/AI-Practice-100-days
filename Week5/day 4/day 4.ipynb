# üóÇ Classification Problems and Common Use Cases

---

## ‚ùì What is Classification?
Classification is a **supervised learning task** where the goal is to **predict discrete class labels (categories)** based on input features.  
- The model learns from **labeled data** and assigns new data points to one or more categories.  
- Widely used in areas like **image recognition, text processing, and medical diagnostics**.

---

## üî¢ Types of Classification

### 1Ô∏è‚É£ Binary Classification
- Predicts **one of two possible classes**.  
- üìå **Example**: Spam vs. Not Spam (email filtering).  

### 2Ô∏è‚É£ Multi-Class Classification
- Predicts **one of more than two classes**.  
- üìå **Example**: Classifying images into **cats, dogs, or birds**.  

### 3Ô∏è‚É£ Multi-Label Classification
- Assigns **multiple labels** to a single instance.  
- üìå **Example**: A news article tagged with multiple topics like **politics, economy, and health**.  

---

# üîë Logistic Regression for Binary Classification

---

## ‚ùì What is Logistic Regression?
Logistic Regression is a **statistical method** used for **predicting binary outcomes** (two possible classes).  
- It estimates the **probability** that a given input point belongs to a **particular class**.  
- Commonly used for **classification tasks** such as spam detection, disease diagnosis, and churn prediction.

---

## üßÆ The Logistic Regression Model

### üìê Equation  
Logistic regression applies the **sigmoid function** to a linear equation:  

\[
P(y=1 \mid x) = \sigma(z) = \frac{1}{1 + e^{-z}}
\]  

Where the linear combination \( z \) is:  

\[
z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
\]  

### üß© Components
- **P(y=1 ‚à£ x)** ‚Äì Probability that the input \(x\) belongs to the **positive class**.  
- **œÉ(z)** ‚Äì Sigmoid (logistic) function that **maps values** between **0 and 1**.  
- **z** ‚Äì Linear combination of **input features** and **model coefficients** \((\beta_0, \beta_1, \dots, \beta_n)\).  

---

## üìå Key Insight
- Outputs a **probability** rather than a direct class label.  
- A **threshold** (commonly 0.5) is applied to assign the final class.  
- Useful for problems where **probabilities** are more informative than binary predictions.  


# üìä Logistic Regression for Binary Classification

---

## üîó The Logistic Regression Model

### üåÄ Sigmoid Function  
- **Purpose:** Maps the output \( z \) to a range between **0 and 1**.  
- **Classification Rule:**  
  \[
  \text{If } P(y=1 \mid x) \geq 0.5, \text{ predict class 1; otherwise, predict class 0.}
  \]  

---

### ‚úÇÔ∏è Decision Boundary  
- **Definition:** The **threshold** used to classify instances.  
- **Default:** **0.5**, but it can be **adjusted**:  
  - Lower threshold ‚Üí **Higher recall** (catch more positives, but risk false positives).  
  - Higher threshold ‚Üí **Higher precision** (fewer false positives, but may miss positives).  

---

### üß≠ Interpretation of Coefficients  
- **Œ≤‚ÇÄ (Intercept):** Baseline log-odds or probability when all features are zero.  
- **Œ≤·µ¢ (Feature Coefficient):**  
  - Represents the effect of feature \(x_i\) on the **log-odds** of the positive class.  
  - Positive Œ≤·µ¢ ‚Üí Increases probability of class 1.  
  - Negative Œ≤·µ¢ ‚Üí Decreases probability of class 1.  


## **Sigmoid Function, Decision Boundary, and Interpretation**

### üìà **Sigmoid Function in Action**
- The **sigmoid function** ensures outputs are **interpretable as probabilities**.  
- A **visualization of the sigmoid function** can illustrate how raw model predictions are mapped to probabilities between **0 and 1**.  

### ‚úÇÔ∏è **Decision Boundary**
- The **default threshold** for classification is **0.5**.  
- **Adjusting the threshold** allows you to **balance precision and recall**, depending on the specific use case or performance requirements.  


```python
import numpy as np

a = -100
b = 100
np.random.seed(42)
X =   a + (b - a) * np.random.rand(100, 1)

y = 2 * X + 3

sigmoid = 1 / (1 + np.exp(-y))
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, sigmoid, color = 'red', label = "Sigmoid based data")
plt.scatter(X, y, color = 'blue', label = " none Sigmoid based data")
plt.legend()
plt.show()
```

```python
#################################### Exercise 1 ####################################
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.model_selection import train_test_split
np.random.seed(42)
n_samples = 200
X = np.random.rand(n_samples, 2) * 10
y = (X[:,0] * 1.5 + X[:, 1] > 15).astype(int)
```

```python
df = pd.DataFrame(X, columns=['Age', 'Salary'])
df['Purchase'] = y
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Age','Salary']], df['Purchase'], test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Acuracy:", accuracy_score(y_test, y_pred))
print("recall_score:", recall_score(y_test, y_pred))
print("f1_score:", f1_score(y_test, y_pred))
print("classification_report:", classification_report(y_test, y_pred))
print("precision_score:", precision_score(y_test, y_pred))
```
