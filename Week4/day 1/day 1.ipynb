# Random Variables

## What are Random Variables?
A **random variable (RV)** is a function that assigns a **number** to each possible outcome of a random experiment.  
This allows us to describe uncertainty in **quantitative** terms.

---

### Example 1: Coin Toss
- Experiment: Toss a coin once. Outcomes = {Head, Tail}.  
- Define a random variable $X$:  
  - $X = 1$ if Head appears  
  - $X = 0$ if Tail appears  

So instead of saying "Head" or "Tail," we can describe the outcome using **numbers**.

---

## Types of Random Variables

### 1. Discrete Random Variables
- Take on **countable values** (finite or infinite).  
- Think of them as outcomes you can **list one by one**.  

#### Real-World Examples
- Number of heads in 10 coin tosses (0, 1, 2, …, 10)  
- Number of customers arriving in a store during 1 hour  
- Number of defective bulbs in a box of 100  

---

#### Probability Mass Function (PMF)
For a discrete random variable $X$, the probability of each possible value $x$ is given by the **PMF**:

$$
P(X = x) = p(x), \quad \sum_x p(x) = 1
$$

- $p(x)$ tells us how likely $X$ takes the value $x$.  
- The sum of all probabilities must equal 1.

---

#### Example 2: 3 Coin Tosses
Let $X =$ number of Heads in 3 tosses of a fair coin.  
Possible values: $X \in \{0,1,2,3\}$  

Probabilities:
- $P(X=0) = \tfrac{1}{8}$ (TTT)  
- $P(X=1) = \tfrac{3}{8}$ (HTT, THT, TTH)  
- $P(X=2) = \tfrac{3}{8}$ (HHT, HTH, THH)  
- $P(X=3) = \tfrac{1}{8}$ (HHH)  

This is an example of a **Binomial distribution**.

---

### 2. Continuous Random Variables
- Take on **uncountably infinite values** (you can’t list them one by one).  
- Think of things measured on a **continuous scale**.  

#### Real-World Examples
- Height of a student (e.g., 170.2 cm, 170.23 cm, …)  
- Time to finish a race (e.g., 10.51 s, 10.512 s, …)  
- Temperature in a city  

---

#### Probability Density Function (PDF)
For a continuous random variable $X$, the probability of $X$ falling inside an interval $[a, b]$ is:

$$
P(a \leq X \leq b) = \int_a^b f(x) \, dx
$$

- Here, $f(x)$ is the **PDF (probability density function)**.  
- Important: $f(x)$ itself is **not a probability**; it’s a "density."  
- The total area under the curve is always 1:

$$
\int_{-\infty}^{\infty} f(x) \, dx = 1
$$

---

#### Example 3: Human Heights
Let $X$ = height of a randomly chosen student.  
Suppose $X \sim \mathcal{N}(170, 10^2)$, i.e. normally distributed with:  
- Mean = 170 cm  
- Standard deviation = 10 cm  

- We can compute probabilities like:  
  $P(160 \leq X \leq 180)$ = area under the PDF curve between 160 and 180.  

This describes the chance that a student’s height lies between 160 and 180 cm.

---

## Expectation and Variance

### Expectation (Mean)
The **expectation** of a random variable is the "long-run average" value if we repeat the experiment many times.  

- For a discrete random variable:

$$
E[X] = \sum_x x \cdot p(x)
$$

- For a continuous random variable:

$$
E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$

---

### Example 4: Die Roll
Let $X$ = outcome of rolling a fair six-sided die.  
- Possible values: 1, 2, 3, 4, 5, 6  
- Each has probability $1/6$  

Expectation:

$$
E[X] = \frac{1+2+3+4+5+6}{6} = 3.5
$$

This means that if you roll the die many times, the average outcome approaches 3.5.

---

### Variance
The **variance** measures how spread out the random variable is around its mean.  

$$
Var(X) = E[(X - E[X])^2]
$$

- A high variance = outcomes are widely spread.  
- A low variance = outcomes are tightly clustered around the mean.  

---

## Summary
- **Discrete RVs**: countable outcomes → described by **PMF**.  
- **Continuous RVs**: uncountable outcomes → described by **PDF**.  
- Both can be summarized by their **expectation** (average outcome) and **variance** (spread).  


```python
from itertools import product

# Sample space of a dice roll
sample_space = list(range(1,7))

# Probability of rolling an even number
even_numbers = [2, 4, 6]
P_even = len(even_numbers) / len(sample_space)
print("P(Even): ", P_even)
```

```python
import numpy as np

# Random variable: dice roll
outcomes = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6] * 6)

# Expectations
excpectation = np.sum(outcomes * probabilities)
print("Expectation (Mean): ", excpectation)

# Variance and Standard Deviation
variance = np.sum((outcomes - excpectation) **2 * probabilities)
std_dev = np.sqrt(variance)

print("Variance: ", variance)
print("Standard Deviation: ", std_dev)
```

# Exercise 1: Simulate Dice Rolls and Calculate Probabilities

In this exercise, we will simulate rolling a fair six-sided die many times and use the results to estimate probabilities.

### Steps:
1. Simulate rolling a die `N` times (e.g., 10000 rolls).  
2. Record the outcomes (1 through 6).  
3. Calculate the relative frequency of each outcome.  
4. Compare the simulated probabilities with the theoretical probability (which should be **1/6 ≈ 0.1667** for each side).


```python
# Exercise 1: Simulate Dice Rolls and Calculate Probabilities

import numpy as np

rolls = np.random.randint(1, 7, size = 10000)
P_even = np.sum(rolls % 2 == 0) / len(rolls)
P_greater_than_4 = np.sum(rolls > 4) / len(rolls)

print("P(Even): ",P_even)
print("P(Greater than 4): ", P_greater_than_4)
```

```python
# Exercise 2: Create and Analyze Random Variables Discrete random variable

import matplotlib.pyplot as plt
import numpy as np

outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6]*6
plt.bar(outcomes, probabilities, color="blue", alpha = 0.7)
plt.title("PMF of a Dice Roll")
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()

```

```python
# Exercise 2: Create and Analyze Random Variables: Continuous random variable
import numpy as np
from scipy.stats import uniform
x = np.linspace(0, 1, 100)
pdf = uniform.pdf(x, loc = 0, scale = 1)
plt.plot(x, pdf, color = "red")
plt.title("PDF of Uniform(0,1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm

x = np.linspace(-3, 3, 300)

# Uniform(0,1)
plt.subplot(1,2,1)
plt.plot(x, uniform.pdf(x, loc=0, scale=1), 'r-', lw=2)
plt.title("Uniform(0,1): Flat PDF")
plt.ylim(0,1.2)

# Normal(0,1)
plt.subplot(1,2,2)
plt.plot(x, norm.pdf(x, loc=0, scale=1), 'b-', lw=2)
plt.title("Normal(0,1): Bell-Shaped PDF")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Parameters
n, p = 50, 0.5
x = np.arange(0, n+1)

# Binomial PMF
pmf = binom.pmf(x, n, p)

# Normal approximation
mu, sigma = n*p, np.sqrt(n*p*(1-p))
x_cont = np.linspace(0, n, 200)
pdf_norm = norm.pdf(x_cont, mu, sigma)

# Plot
plt.bar(x, pmf, alpha=0.6, label="Binomial PMF")
plt.plot(x_cont, pdf_norm, 'r-', lw=2, label="Normal Approximation")
plt.xlabel("k (number of successes)")
plt.ylabel("Probability")
plt.title("Binomial vs Normal Approximation")
plt.legend()
plt.show()
```
