# üîó Understanding Correlation

---

## üìñ **What is Correlation?**
Correlation measures the **strength** and **direction** of the relationship between **two variables**.  
It helps determine **how closely** changes in one variable are associated with changes in another.

---

## üìä **Key Points**
- **Values range:** ‚Äì1 to 1  
  - **+1** ‚Üí Perfect positive correlation (as one variable increases, the other increases proportionally).  
  - **0** ‚Üí No correlation (no linear relationship).  
  - **‚Äì1** ‚Üí Perfect negative correlation (as one variable increases, the other decreases proportionally).  
- Correlation **does not imply causation** ‚Äî two variables may be correlated due to coincidence or a hidden factor.

---

## üß∞ **Types of Correlation**

### 1Ô∏è‚É£ **Pearson Correlation Coefficient (\(r\))**
- Measures the **linear relationship** between two continuous variables.  
- Sensitive to **outliers**.  
- Formula:  
  $$
  r = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}
  $$  
  where:
  - \(\text{Cov}(X,Y)\) = covariance between \(X\) and \(Y\),  
  - \(\sigma_X, \sigma_Y\) = standard deviations of \(X\) and \(Y\).

---

### 2Ô∏è‚É£ **Spearman Rank Correlation Coefficient (\(\rho\))**
- Measures the **monotonic relationship** (does not require linearity).  
- Uses **ranked values** instead of raw data ‚Üí **robust to outliers** and works with ordinal data.  
- Formula:  
  $$
  \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}
  $$  
  where:
  - \(d_i\) = difference between ranks of corresponding values,  
  - \(n\) = number of pairs.

---

## üß† **Example Use Cases**
- **Pearson**: Relationship between **height** and **weight**.  
- **Spearman**: Relationship between **movie rankings** by two critics.  

---

## üêç **Python Implementation**
```python
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Example data: Study hours vs. Exam scores
study_hours = np.array([2, 4, 5, 7, 9, 10])
exam_scores = np.array([50, 55, 65, 70, 80, 85])

# Pearson Correlation
pearson_corr, pearson_p = pearsonr(study_hours, exam_scores)
print(f"Pearson r: {pearson_corr:.3f}, p-value: {pearson_p:.4f}")

# Spearman Correlation
spearman_corr, spearman_p = spearmanr(study_hours, exam_scores)
print(f"Spearman œÅ: {spearman_corr:.3f}, p-value: {spearman_p:.4f}")


````python
import numpy as np
from scipy.stats import pearsonr, spearmanr

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

r, _ = pearsonr(x,y)
print("Pearson Correlation Coefficient", r)

rho, _ = spearmanr(x, y)
print("Spearman Correlation Coefficient: ", rho)

````

# üìà Linear Regression

---

## üìñ **What is Linear Regression?**
Linear Regression is a **statistical method** used to model the relationship between:
- A **dependent variable** (\(y\)) ‚Äî the outcome we want to predict.
- One or more **independent variables** (\(x\)) ‚Äî the predictors or features.

It finds the **best-fitting straight line** that minimizes the difference between predicted and actual values.

---

## üßÆ **Formula for Simple Linear Regression**
\[
y = \beta_0 + \beta_1 x + \varepsilon
\]

Where:  
- \(\beta_0\) = Intercept (value of \(y\) when \(x = 0\))  
- \(\beta_1\) = Slope (how much \(y\) changes for a one-unit change in \(x\))  
- \(\varepsilon\) = Error term (difference between observed and predicted values)  

---

## üîë **Key Points**
- **Goal**: Find values of \(\beta_0\) and \(\beta_1\) that minimize the **sum of squared errors (SSE)**.  
- **Assumptions**:
  1. **Linearity**: Relationship between \(x\) and \(y\) is linear.  
  2. **Independence**: Observations are independent of each other.  
  3. **Homoscedasticity**: Constant variance of errors.  
  4. **Normality**: Errors are normally distributed.  

---

## üß† **Use Cases**
- Predicting **house prices** based on square footage.  
- Estimating **sales revenue** based on advertising spend.  
- Modeling **student grades** based on study hours.  

---

## üêç **Python Implementation**
```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example data: Study hours vs Exam scores
X = np.array([2, 4, 5, 7, 9, 10]).reshape(-1, 1)  # Independent variable (study hours)
y = np.array([50, 55, 65, 70, 80, 85])           # Dependent variable (exam scores)

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Get slope and intercept
beta_1 = model.coef_[0]
beta_0 = model.intercept_

print(f"Intercept (Œ≤‚ÇÄ): {beta_0:.2f}")
print(f"Slope (Œ≤‚ÇÅ): {beta_1:.2f}")

# Predict values
y_pred = model.predict(X)

# Plot
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Linear Regression Example")
plt.legend()
plt.show()


````python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 8, 10])

# Fit Linear Regression
model = LinearRegression()
model.fit(x, y)

print("Slope: ", model.coef_[0])
print("Intercept: ", model.intercept_)
print("R-Squared: ", model.score(x, y))
````

````python
######################################### Exercise 2 #########################################
import numpy as np 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load Iris dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/iris.csv"
df = pd.read_csv(url)
del df["species"]

x = np.array(df["petal_length"]).reshape(-1, 1)
y = np.array(df["petal_width"])

model = LinearRegression()
model.fit(x, y)

print("Slope: ", model.coef_[0])
print("Intercept: ", model.intercept_)
print("R-Squared: ", model.score(x, y))

plt.scatter(x, y, color="blue", label = "Data")
plt.plot(x, model.predict(x), color = "red", label = "Regression Line")
plt.legend()
plt.title("Linear Regression")
plt.show()
````
