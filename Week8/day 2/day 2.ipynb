# üîÑ Understanding Forward Propagation

---

## üìå What is Forward Propagation?

**Forward Propagation** is the process by which **input data flows through the layers** of a neural network ‚Äî from the input layer to the output layer ‚Äî to generate predictions.

Each layer transforms its input using **weights**, **biases**, and an **activation function** before passing the result to the next layer.

Mathematically:
\[
z = \sum_{i=1}^{n} (w_i \times x_i) + b
\]
\[
a = f(z)
\]
Where:  
- \( x_i \): input feature  
- \( w_i \): weight for input \( x_i \)  
- \( b \): bias term  
- \( f \): activation function  
- \( a \): output (activation) sent to the next layer  

---

- Passes these inputs directly to the next layer.

## üìå Overview

**Forward Propagation** involves a sequence of mathematical operations performed in each layer of a neural network to transform inputs into outputs (predictions).  
It‚Äôs the ‚Äúforward pass‚Äù of the network ‚Äî where data moves from **input ‚Üí hidden ‚Üí output** layers.

---

## üßÆ Step 1: Compute Weighted Sum

Each neuron calculates a **weighted sum** of its inputs plus a **bias term**.

\[
z = W \cdot X + b
\]

Where:  
- \( W \): Weight matrix (importance of each input)  
- \( X \): Input vector  
- \( b \): Bias (shifts activation threshold)

This linear transformation combines all inputs for each neuron.

# ‚ö° Common Activation Functions

---

## üìå Introduction

Activation functions introduce **non-linearity** into a neural network, enabling it to learn complex patterns and relationships.  
Without them, even a multi-layer neural network would behave like a **linear regression model**.

---

## üß† 1. Sigmoid Function

\[
\sigma(z) = \frac{1}{1 + e^{-z}}
\]

### ‚úÖ **Use Case**
- Commonly used in **binary classification** for the output layer.

### ‚ö†Ô∏è **Limitation**
- Suffers from **vanishing gradients** for very large or very small \( z \).  
  ‚Üí Gradients approach zero, slowing or stopping learning.

### üíª **Example**

# üéØ Choosing Activation Functions

---

## üìå Why Activation Function Choice Matters

Selecting the right activation function is crucial for **model stability**, **learning efficiency**, and **accuracy**.  
Different layers and tasks require specific types of activation functions to balance **gradient flow** and **output behavior**.

---

## üß≠ Recommended Activation Functions by Layer Type

| **Layer Type** | **Recommended Activation Function** | **Rationale** |
|----------------|--------------------------------------|----------------|
| **Hidden Layers** | ReLU or Tanh | Efficient gradient flow, reduces vanishing gradients |
| **Output (Binary Classification)** | Sigmoid | Outputs probability between 0 and 1 |
| **Output (Multi-Class Classification)** | Softmax | Converts logits to normalized probability distribution |
| **Output (Regression)** | None / Linear | Keeps continuous numerical outputs without squashing |

---

## ‚öôÔ∏è Guidelines for Choosing

### üîπ Hidden Layers
- **Preferred:** `ReLU`
  - Efficient, fast convergence, sparse activations.
- **Alternative:** `Tanh`
  - Zero-centered output; useful when gradients oscillate.

````python
import numpy as np

#Define activation functions

def sigmoid(z):
    return 1/ (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)

def softmax(z):
    exp_z = np.exp(z- np.max(z))
    return exp_z / exp_z.sum(axis = 0, keepdims = True)

# Forward pass function
def forward_pass(X, weights, biases, activation_function):
    z = np.dot(weights, X) + biases
    a = activation_function(z)
    return a

X = np.array([[0.5], [0.8]])
weights = np.array([[0.2, 0.4], [0.6, 0.1]])
biases = np.array([[0.1], [0.2]])

print(forward_pass(X, weights, biases, sigmoid))
print(forward_pass(X, weights, biases, tanh))
print(forward_pass(X, weights, biases, relu))
print(forward_pass(X, weights, biases, softmax))
````

````python
import matplotlib.pyplot as plt

# Define range of inputs
z = np.linspace(-10, 10, 100)

# Plot activation functions
plt.figure(figsize=(12, 8))
plt.plot(z, sigmoid(z), label = "Sigmoid")
plt.plot(z, tanh(z), label = "tanh")
plt.plot(z, relu(z), label = "relu")
plt.plot(z, softmax(z), label = "softmax")
plt.title("Activation Functions")
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.legend()
plt.grid(True)

plt.show()
````
