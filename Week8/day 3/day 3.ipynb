# üìâ Understanding Loss Functions

---

## üìå What Are Loss Functions?

A **loss function** (also called a **cost function** or **objective function**) quantifies how far a model‚Äôs predictions are from the actual target values.  
It is the **core feedback mechanism** during training ‚Äî telling the model *how wrong* it is and *how to improve*.

---

## üéØ Key Purposes of Loss Functions

### 1. **Error Measurement**
- Measures the **difference** between predicted outputs (\( \hat{y} \)) and actual targets (\( y \)).
- Indicates how accurate the model‚Äôs predictions are.

\[
\text{Loss} = f(y, \hat{y})
\]

### 2. **Feedback for Optimization**
- Guides the learning process by providing a signal for **backpropagation**.
- The network computes **gradients** of the loss with respect to weights and biases and updates them to minimize the loss.

\[
\text{Goal: } \min_{\theta} \mathcal{L}(y, \hat{y})
\]

Where \( \theta \) are model parameters (weights, biases).

---

## üß© Role in Neural Networks

| Role | Description |
|------|--------------|
| **Error Measurement** | Quantifies prediction accuracy or deviation. |
| **Optimization Guidance** | Provides direction for parameter updates during training. |
| **Training Objective** | Defines what the model learns to optimize. |

---

## üîç How It Works in Practice

1. The model makes predictions \( \hat{y} \).  
2. The loss function computes the error:  
   \[
   L = f(y, \hat{y})
   \]
3. The optimizer (e.g., Gradient Descent) uses this loss to adjust model weights.  
4. This process repeats over many iterations (epochs).

---

## üí° Example: Mean Squared Error (Regression)

The **Mean Squared Error (MSE)** loss measures the average squared difference between predicted and actual values.

\[
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
\]

### üíª Example in Code
```python
import numpy as np

# Actual vs predicted values
y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5, 0.0, 2.1, 7.8])

# Compute MSE
mse = np.mean((y_true - y_pred)**2)
print("Mean Squared Error:", mse)


# üßÆ Understanding Loss Functions ‚Äî Common Types

---

## üìå Overview

Loss functions play a key role in guiding a neural network‚Äôs learning process.  
Different tasks require different loss functions, depending on whether the output is **continuous** (regression) or **categorical** (classification).

---

## üîπ 1. Mean Squared Error (MSE)

### üìò Definition
The **Mean Squared Error (MSE)** is a common loss function for **regression** problems.  
It measures the average squared difference between predicted values (\( \hat{y} \)) and actual target values (\( y \)).

\[
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
\]

### üí° Key Characteristics
- Penalizes **larger errors more heavily** (squared term).  
- Smooth and differentiable ‚Äî good for optimization via gradient descent.  
- Sensitive to **outliers**.

### üß© Use Case
Used when the target variable is continuous (e.g., predicting house prices, temperature, stock value).

---

### üíª Example in Python
```python
import numpy as np

# True vs predicted values
y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5, 0.0, 2.1, 7.8])

# Compute MSE
mse = np.mean((y_true - y_pred)**2)
print("Mean Squared Error:", mse)


## üîπ Cross-Entropy Loss

---

### üìò Definition

The **Cross-Entropy Loss** (also known as **Log Loss**) is used for **classification tasks**.  
It measures the difference between the **true labels** (\( y \)) and the **predicted probabilities** (\( \hat{y} \)).

\[
\text{Cross-Entropy} = -\frac{1}{n} \sum_{i=1}^{n} \Big[ y_i \log(\hat{y}_i) + (1 - y_i)\log(1 - \hat{y}_i) \Big]
\]

For **multi-class classification** (Softmax output):

\[
\text{CCE} = -\sum_{i=1}^{C} y_i \log(\hat{y}_i)
\]

Where \( C \) = number of classes.

---

### üí° Key Characteristics

- ‚úÖ Suitable for **probabilistic outputs** (from **Sigmoid** or **Softmax**).  
- ‚ö†Ô∏è Penalizes **confident but incorrect predictions** strongly.  
- üéØ Encourages the model to assign **high probability to the correct class**.

---

### üß© Use Cases

- **Binary Cross-Entropy (BCE)** ‚Üí Binary classification (e.g., spam vs. not spam).  
- **Categorical Cross-Entropy (CCE)** ‚Üí Multi-class classification (e.g., digit recognition, sentiment analysis).

---

### üíª Example: Binary Cross-Entropy

```python
import numpy as np

# True labels and predicted probabilities
y_true = np.array([1, 0, 1, 0])
y_pred = np.array([0.9, 0.2, 0.8, 0.1])

# Compute Binary Cross-Entropy
bce = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
print("Binary Cross-Entropy Loss:", bce)


# üîÅ Introduction to Backpropagation

---

## üìò What Is Backpropagation?

**Backpropagation** (short for *backward propagation of errors*) is the process of **computing gradients** for each weight and bias in a neural network.  
It enables **optimization algorithms** (like *Gradient Descent*) to adjust parameters and **minimize the loss function**.

Essentially, it allows the network to *learn from its mistakes* ‚Äî by measuring how much each parameter contributed to the total error.

---

## ‚öôÔ∏è Steps in Backpropagation

1. **Forward Pass**
   - Compute the output and loss using the current weights and biases.
   - Data flows from the input layer to the output layer.
   - Example: \( \hat{y} = f(Wx + b) \)

2. **Backward Pass**
   - Calculate the **gradient** of the loss function with respect to each parameter.
   - Uses the **chain rule** of calculus to propagate the error backward through the layers.

3. **Weight Update**
   - Update weights using the computed gradients.
   - Adjust in the opposite direction of the gradient to minimize the loss:
     \[
     W_{\text{new}} = W_{\text{old}} - \eta \frac{\partial L}{\partial W}
     \]
     where \( \eta \) is the **learning rate**.

---

## üß† Key Concepts

- **Gradient:**  
  The rate of change of the loss with respect to a parameter  
  \[
  \text{Gradient} = \frac{\partial L}{\partial W}
  \]

- **Gradient Descent:**  
  An optimization algorithm that minimizes the loss by updating parameters in the direction of the **negative gradient**.

---

## üí° Intuition

Think of backpropagation like **blame assignment** in a team:
- Each neuron contributes differently to the final prediction.
- Backpropagation identifies *how much each neuron‚Äôs parameters* (weights & biases) are responsible for the final error.
- Then it adjusts those parameters accordingly.

---

## üíª Example (Simple Neural Network Training)

```python
import numpy as np

# Example data
x = np.array([0.5])
y_true = np.array([1.0])

# Initialize weight and bias
W = np.random.randn()
b = np.random.randn()
lr = 0.1  # learning rate

# Forward pass
y_pred = W * x + b
loss = (y_true - y_pred)**2  # MSE Loss

# Backward pass (compute gradients)
dL_dy = -2 * (y_true - y_pred)
dL_dW = dL_dy * x
dL_db = dL_dy * 1

# Update weights and bias
W -= lr * dL_dW
b -= lr * dL_db

print(f"Updated Weight: {W:.4f}, Updated Bias: {b:.4f}")


````python
import numpy as np

# Mean Squared Error(MSE) Loss

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# $ Binary Cross-Entropy (BCE) loss 
def bce_loss(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-15, 1-1e-15)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1-y_pred))

# Example Data
y_true = np.array([1, 0, 1, 1])
y_pred = np.array([0.9, 0.2, 0.8, 0.7])

# Calculate loss
mse = mse_loss(y_true, y_pred)
bce = bce_loss(y_true, y_pred)

print(mse, bce)
````

````python
# Derivative of MSE loss
def mse_gradient(y_true, y_pred):
    return 2 * (y_pred - y_true) / len(y_true)

# Derivative of BCE loss
def bce_gradient(y_true, y_pred):
    y_pred = y_pred = np.clip(y_pred, 1e-15, 1-1e-15)
    return (y_pred-y_true) / (y_pred * (1-y_pred))

# Calculate gradients
grad_mse = mse_gradient(y_true, y_pred)
grad_bce = bce_gradient(y_true, y_pred)

print(grad_mse, grad_bce)
````

````python
import matplotlib.pyplot as plt

# Define predictions and true labels
predictions = np.linspace(0, 1, 100)
true_label = 1

mse_loss = [(true_label - p) ** 2 for p in predictions]
bce_loss = [-true_label * np.log(max(p, 1e-15)) - (1 - true_label) * np.log(max(1-p,1e-15)) for p in predictions]

plt.figure(figsize=(11,7))
plt.plot(predictions, mse_loss, label = "MSE")
plt.plot(predictions, bce_loss,  label = "BCE")
plt.xlabel('preds')
plt.ylabel('losses')
plt.legend()
plt.show()
````
